```python
import cv2
import os
import pandas as pd
import sklearn
import sklearn.metrics
import random
import matplotlib.pyplot as plt
```

```python
def get_confusion_matrix(file):
    file = pd.read_csv(file)
    file = file.set_index(['Unnamed: 0'])
    return file
confusion_matrix = get_confusion_matrix('confusion_matrix.csv')
```

```python
import seaborn as sn
import numpy as np
sn.heatmap(np.log(confusion_matrix),annot=True)
```

```python
def get_data(pred_loc,data_loc=''):
    pred = cv2.imread(pred_loc)
    pred *= 40 # As values, are small it will help, to see them better
    target = cv2.imread(data_loc)
    target *= 40
    plt.imshow(np.concatenate([pred, target], 1))
    plt.show()
random_files = random.sample(os.listdir('output2/'),10)
for i in random_files:
    get_data('output2/' + i,'target/' + i)
    
```

```python
def get_specific_class(idx,from_,n=10):
    imgs = []
    for i in range(n):
        while True:
            addr = random.sample(os.listdir(from_),1)[0]
            img = cv2.imread(from_ + addr)
            if idx in np.unique(img):
                imgs.append(addr)
                break
    return imgs
samples = get_specific_class(1,'target/',10)
print(samples)
```

```python
samples = get_specific_class(1,'target/',5)
for i in samples:
    get_data('output2/' + i,'target/' + i)
```
